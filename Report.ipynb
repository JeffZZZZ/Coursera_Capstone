## --- Machine Learning Model for Neighbourhoods in Toronto

## 1. Introduction

This is an era of peace and plenty. Most of people never suffer from security threat or hunger, the basic needs are also be satisfied. Thus, the happiness of life significantly depends on some higher level deisres such as more delicate food and drink, diversified entertainment and deeper education. These indicators can be quantificated as the number of restaurants, entertainment facilities and schools.
<br>
The objective of this project is to build a machine learning model to cluster the neighbourhoods of Toronto, count the number of venues, and find out the area where people will be the most happy to live in.
<br>
The potential users would be everyone that is interested to move to Ottawa and begin their new life.

## 2. Data

The postal code, borough and neighbourhood names can be found on Wikipeida: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M
<br>
The coordinate of each postal code can be found here: http://cocl.us/Geospatial_data
<br>
The venues data is gained from: https://foursquare.com

## Methodology

List of tools:
- Cloud computing platform: IBM Watson Studio
- Editor: Jupyter Notebook
- Programming language: Python 3.6
- Location data source: Foursquare.com
<br>
First, import data into python pandas dataframe and preprocess the table. Then use folium library to visulize the neighbourhoods of Toronto, and use location data gained from foursquare.com to merge the venues infomation. Finally, use k-means clustering algorithm to cluster the neighbourhoods and analyze the results.
