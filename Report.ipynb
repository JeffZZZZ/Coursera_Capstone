## 1. Introduction

This is an era of peace and plenty. Most of people never suffer from security threat or hunger, the basic needs are also be satisfied. Thus, the happiness of life significantly depends on some higher level deisres such as more delicate food and drink, diversified entertainment and The amount of outdoor recreation spots. These indicators can be quantificated as the number of restaurants, entertainment facilities and outdoor recreation areas.
<br>
The objective of this project is to build a machine learning model to cluster the neighbourhoods of Ottawa, count the number of venues, and find out the area where people will be the most happy to live in. This is a preliminary and conceptual research that contains only the basic data and machine learning model, which demonstrates a possible direction for future deeper research.
<br>
The potential users would be anyone that is interested to move to Ottawa and start a new life.

## 5. Discussion

As we can see from the above bar chart in Result section, cluster 3 has the most amount of all kinds of venues in Ottawa. These neighbourhoods have hundreds of restaurants, and dozens of entertainment facilities and outdoor or recreation areas. Cluster 1 is very similar to cluster 3 but lacks a small amout of restaurants and entertainments. Cluster 2 is the medium of all 5 clusters, while cluster 0 has the least amount of venues. 
<br>
In all clusters, restaurant has the most occurence while entertainment facility has the least amount. It is interesting that all categories of venues have the same comparison relation among the clusters, which means if one category of the venues in one cluster is less than the other cluster, the other categories of venues must also be less. 
<br>
Overall, the prosperity comparison among all clusters is: 3 > 1 > 2 > 4 > 0, or in the map: Green > Purple > Blue > Orange > Red
<br>
It also shows the development of Ottawa. This city starts at the old Ottawa area closed to Ottawa River, and then expands outwards. This is why the prosperity trend is outspread.

## 6. Conclusion

In conclusion, the best neighbourhoods that can provide the best life quality is in cluster 3, which are ByWard Market, Centretown,	Downtown, Lower Town, Old Ottawa East, Sandy Hill, Wellington Village, Kempark and South Gloucester. Neighbourhoods that fall in cluster 1 are also very good places to live. However, other neighbourhoods have significantly less venues than the first two. Finally, people will have a very boring life if they live in cluster 0.
